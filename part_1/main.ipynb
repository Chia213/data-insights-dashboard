### 21 Använd .apply() 
1. Undersök vilka kategorier det finns i kolumnen "state"
2. Skapa en funktion som baserat på värdena i kolumnen state är, modifierar kolumnen "usd_funding_goal" med en * om det står success, "!" för failure eller canceled, "---" för "suspended".
3. Displaya ett sample av 20 rader från din modifierade kolumn  

```python
def change_usd_funding_goal_to_symbol(status):
  if status["state"] == "succesful":
    return status["usd_funding_goal"] + "*"
  elif status["state"] == "failed":
    return status["usd_funding_goal"] + "!"
  elif status["state"] == "canceled":
    return status["usd_funding_goal"] + "!"
  elif status["state"] == "suspended":
    return status["usd_funding_goal"] + "---"

df["usd_funding_goal"] = df["usd_funding_goal"].astype(str)
df["usd_funding_goal"] = df.apply(change_usd_funding_goal_to_symbol, axis = 1)
df["usd_funding_goal"].head(20)
#df.loc[df["pledged"] == 0.0, ["pledged"]] = np.nan
#df.loc[df["state"] == "successful", ["usd_funding_goal"]] = "*"
```
